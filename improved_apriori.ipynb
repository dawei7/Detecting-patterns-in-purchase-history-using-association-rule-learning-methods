```python
my_dict = dict()

def modified_sigmoid(x):
    return 1 / (1 + math.exp(-x+4))

date_diff = (data["date"].max()-data["date"].min()).days

for i in range(date_diff):
    my_dict[data["date"].min()+timedelta(days = i)]=modified_sigmoid(i/date_diff*8)
```
